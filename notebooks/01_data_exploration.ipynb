# 01 — Data Exploration

Run this AFTER you've downloaded CIC-IDS2017 CSVs into `data/raw/`.

Goals:
- Verify each file loads cleanly
- Check class distribution
- Inspect feature distributions
- Confirm there are enough samples per class for 500/class experiments

```
import sys; sys.path.insert(0, '..')
import pandas as pd
import matplotlib.pyplot as plt
from src.data.loader import load_raw, class_counts
from src.data.feature_blacklist import get_blacklist
```

```
df_ddos = load_raw('friday_ddos')
print('Friday DDoS:', df_ddos.shape)
print(class_counts(df_ddos))
```

```
df_ps = load_raw('friday_portscan')
print('Friday PortScan:', df_ps.shape)
print(class_counts(df_ps))
```

```
df_dos = load_raw('wednesday_dos')
print('Wednesday DoS:', df_dos.shape)
print(class_counts(df_dos))
```

```
# Confirm we have enough samples for 500/class experiments
for name, df in [('DDoS', df_ddos), ('PortScan', df_ps), ('DoS', df_dos)]:
    counts = df['binary_label'].value_counts()
    enough = (counts >= 500).all()
    print(f'{name}: BENIGN={counts.get(0, 0)}, ATTACK={counts.get(1, 0)}, 500/class OK: {enough}')
```

```
# List features after blacklist removal
blacklist = set(get_blacklist())
feature_cols = [c for c in df_ddos.columns if c not in blacklist]
print(f'{len(feature_cols)} usable features:')
for c in feature_cols:
    print(' -', c)
```
